```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('StudentsPerformance-modified1.csv')
df
```

```python
#Scanning all inconsistencies
df.isnull().sum()
```

```python
df=df.dropna(subset=['reading score'])#coloumn wise removal of NULL values.
```

```python
df.dropna(axis=0)
```

```python
df['math score'].fillna(df['math score'].mean(), inplace=True)
```

```python
df['math score'].head(20)
```

```python
df['math score'].fillna(df['math score'].median(), inplace=True)
df['math score'].head(25)
```

```python
df['math score'].median()
```

```python
df['math score'].fillna(df['math score'].mode().iloc[0], inplace=True)
df['math score'].mode()
```

```python
df['math score'].head(25)
```

```python
maxx=df['math score'].max()#min-max norm. method
minn=df['math score'].min()
dfn=[((x-minn)/(maxx-minn)) for x in df['math score']]
df['math_score_norm']=dfn
#df.head()
#df['math score'].plot(kind='box')
df['math_score_norm'].plot(kind='box')
```

```python
meann=df['reading score'].mean()#z-score method
stdd=df['reading score'].std()
dfz=[((x-meann)/stdd) for x in df['reading score']]
df['reading_score_z']=dfz;
df.head(10)
```

```python
q1 = df.quantile(0.25)#IQR and box-plot method
q3 = df.quantile(0.75)

iqr=q3-q1
```

```python
ll = q1-(1.5*iqr)#formulas
ul = q3+(1.5*iqr)
```

```python
ll
```

```python
ul
```

```python
low=df<ll
high=df>ul
```

```python
high
```

```python
df[high | low]#Union of lower & higher values than ll and ul 
```

```python
out=df[~(low | high).any(axis=1)]
out.plot(kind="box")
```
